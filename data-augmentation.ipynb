```python
import os
os.makedirs('/kaggle/working/train/absent')
os.makedirs('/kaggle/working/train/present')
os.makedirs('/kaggle/working/test/absent')
os.makedirs('/kaggle/working/test/present')
os.makedirs('/kaggle/working/valid/absent')
os.makedirs('/kaggle/working/valid/present')
```

```python
from keras.preprocessing.image import ImageDataGenerator

from tensorflow.keras.utils import img_to_array, array_to_img, load_img
datagen= ImageDataGenerator(
    width_shift_range= 0.2,
    height_shift_range= 0.2,
    shear_range= 0.2,
    fill_mode= 'nearest',
    horizontal_flip= True,    
    zoom_range= 0.2
)
```

```python
from tqdm import tqdm
def augment(i_path, o_path):
    for img in tqdm(os.listdir(i_path)):
        x=os.path.join(i_path,img)
        img= load_img(x)
        x= img_to_array(img)
        x= x.reshape((1,)+ x.shape)
        i=0
        for batch in datagen.flow(x, batch_size=1, save_to_dir=o_path, save_format='png'):
            i=i+1
            if i>10:
                break
```

```python
i_v_a='/kaggle/input/xraykneejoint/Valid/absent'
i_v_p='/kaggle/input/xraykneejoint/Valid/present'
i_tr_a='/kaggle/input/xraykneejoint/train/absent'
i_tr_p='/kaggle/input/xraykneejoint/train/present'
i_te_a='/kaggle/input/xraykneejoint/test/absent'
i_te_p='/kaggle/input/xraykneejoint/test/present'


o_v_a='/kaggle/working/valid/absent'
o_v_p='/kaggle/working/valid/present'
o_tr_a='/kaggle/working/train/absent'
o_tr_p='/kaggle/working/train/present'
o_te_a='/kaggle/working/test/absent'
o_te_p='/kaggle/working/test/present'
```

```python
augment(i_v_a, o_v_a)
augment(i_v_p, o_v_p)
```

```python
augment(i_tr_a, o_tr_a)
augment(i_tr_p, o_tr_p)
```

```python
augment(i_te_a, o_te_a)
augment(i_te_p, o_te_p)
```
